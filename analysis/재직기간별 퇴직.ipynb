```python
import pandas as pd
data_재직기간별_퇴직 = pd.read_csv("./data/사립학교교직원연금공단_재직기간별 학교급별 교원 퇴직 현황_20200916.csv",encoding = 'euc-kr')
```

```python
data_재직기간별_퇴직
```

```python
df=data_재직기간별_퇴직
```

```python
for a in range(0,104):
    if df["재직기간"][a]=="5년 미만":
        df['재직기간'][a]=df['재직기간'][a].replace("5년 미만",'00년 이상~05년 미만')
    elif df["재직기간"][a]=="5년 이상~10년 미만":
        df['재직기간'][a]=df['재직기간'][a].replace("5년 이상~10년 미만",'05년 이상~10년 미만')
    else:
        continue
```

```python
df.head()
```

```python
df_sorted_by_values = df.sort_values(by=['재직기간', '연도'] ,ascending=True)
```

```python
df_sorted_by_values.head()
```

```python
df_sorted_by_values = df_sorted_by_values.reset_index(drop=True)
```

```python
df_sorted_by_values
```

```python
del df_sorted_by_values['연도']
```
