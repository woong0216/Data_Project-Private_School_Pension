```python
import pandas as pd
data = pd.read_excel("./data/기간별 재직기간에 따른 퇴직연금.xlsx")
```

```python
data
```

```python
data_2=data.loc[:,["퇴직연금 지급액(천원)","유치원 및 초등학교","중학교","고등학교","전문대학","대학(교)","특수학교"]]
```

```python
data_2.corr()
```

```python
data_2
```

```python
# 데이터프레임 설정 x y 설정
X, y = data_2.iloc[:,1:], data_2.iloc[:,0] # X는 1~27행에서 2~6열  Y는 1~27행부터 1열
print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#모델 객체 생성
m_reg = LinearRegression(fit_intercept = True) #fit_intercept -> beta0 생성 여부

#모델 학습
m_reg.fit(X, y) # 선형회귀 분석 툴 생성
```

```python
# b0값
m_reg.intercept_
```

```python
# x들의 계수
m_reg.coef_
```

```python
# 선형 모델의 절편값 β0 # 추정된 상수항
beta0 = m_reg.intercept_

# 선형 모델의 기울기값 β1 # 추정된 가중치 벡터
beta1 = m_reg.coef_
print('y = {} + {} x1 + {} x2 + {} x3 + {} x4 + {} x5 + {} x6'.format(round(beta0,3), round(beta1[0],3), round(beta1[1],3),\
                                                    round(beta1[2],3), round(beta1[3],3), round(beta1[4],3),round(beta1[5],3)))
```

```python
beta1[5]
```

```python
import numpy as np
```

```python
y_pred = m_reg.predict(X)
print(np.round(list(y_pred),3)) # 27개
print(np.round(m_reg.score(X, y),3))
```

```python
from sklearn.metrics import r2_score as r2
r2 = np.round(r2(y, y_pred),3)
r2
```
